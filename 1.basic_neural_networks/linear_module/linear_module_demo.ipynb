```python
import torch
import torch.nn as nn
```

Make a linear module that takes an input size of 5 and returns output size of 3

```python
mod= nn.Linear(5,3)
print(mod)
```

lets make a random tensor of size 5

```python
x=torch.rand(5)
print(x)
print(x.size())
```

Feed it to the model

```python
y=mod(x)
print(y)
```

```python
print(mod.weight)
print(mod.weight.size())
print(mod.bias)
print(mod.bias.size())
```

We can also make a linear model without bias

```python
mod2=nn.Linear(5,3,bias=False)
print(mod2)
```

```python
print(mod2.weight)
print(mod2.bias)
```
