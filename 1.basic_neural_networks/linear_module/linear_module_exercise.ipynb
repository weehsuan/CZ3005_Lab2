```python
import torch
import torch.nn as nn
import torch.nn.functional as F 
```

Make a linear module without bias that takes an input size of 2 and return an output size of 3

```python
mod=nn.Linear(2,3,bias=False)
print(mod)
```

Print the internal parameters

```python
print(mod.weight)
print(mod.bias)
```

make a vector and feed it to your linear module

```python
x=torch.Tensor([1,1])
print(x)
y=mod(x)
print(y)
```
