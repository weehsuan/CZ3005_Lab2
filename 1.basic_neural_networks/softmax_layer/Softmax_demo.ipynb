```python
import torch
import torch.nn.functional as F
```

Make a vector

```python
x=torch.Tensor([-2, -0.5, 2, 1.5])
print(x)
print(x.size())
```

Feed it to the softmax

```python
p=F.softmax(x,dim=0)
print(p)
print(p.size())
```

Check that it sums to one

```python
print(p.sum())
```

make a matrix

```python
A=torch.tensor([[-2,-0.5,2.0,1.5],[-2,-0.5,7.0,1.5]])
print(A)
```

*Take* the softmax over the row (dim=1)

```python
p=F.softmax(A,dim=1)
print(p)
```

Take the softmax over the column (dim=0)

```python
p=F.softmax(A,dim=0)
print(p)
```
