```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

In pytorch, networks are defined as a class

```python
class mynet(nn.Module):

  def __init__(self,input_size,hidden_size,output_size):
    super(mynet, self).__init__()

    self.layer1=nn.Linear(input_size,hidden_size,bias=True)
    self.layer2=nn.Linear(hidden_size,output_size,bias=True)

  def Forward(self,x):

    x=self.layer1(x)
    x=F.relu(x)
    x=self.layer2(x)
    p=F.softmax(x,dim=0)

    return p
```

create something that takes an input of size 2, then transforms it into something size of 5, then into something size of 3

```python
net = mynet(2,5,3)
print(net)
```

now we are going to make an input vector and feed it to the network

```python
x=torch.Tensor([1,1])
print(x)
p=net.Forward(x)
print(p)
print(p.sum())
```

We can access the first layer as follows

```python
print(net.layer1)
```

We can get the network weight and bias

```python
print(net.layer1.weight)
print(net.layer1.bias)
print(net.layer2.weight)
print(net.layer2.bias)
```

alternatively, all the network parameters can be accessed by using net.parameters()

```python
list_of_params=list(net.parameters())
print(list_of_params)
```
