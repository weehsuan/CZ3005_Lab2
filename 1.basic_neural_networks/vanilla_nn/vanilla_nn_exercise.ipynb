```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

Lets create one layer neural network with bias

```python
class mynet(nn.Module):
  def __init__(self,input_size,output_size):
    super(mynet,self).__init__()

    self.mylayer=nn.Linear(input_size,output_size,bias=True)

  def forward(self,x):

    x=self.mylayer(x)
    p=F.softmax(x,dim=0)

    return p
```

Create an instance of a one layer net taking input of size 2 and returning output size of 2

```python
net=mynet(2,2)
print(net)
```

print the parameters of the neural network

```python
print(net.mylayer.weight)
print(net.mylayer.bias)
```

make a vector and feed it to the network

```python
x=torch.Tensor([1,1])
print(x)
p=net(x)
print(p)
print(p.sum())
```
